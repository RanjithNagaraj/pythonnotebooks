```python
import seaborn as sns
%matplotlib inline
```

```python
#loading inbuilt dataset provided by seaborn for practising vis plots
tips = sns.load_dataset("tips")
```

```python
tips.head(10)
```

```python
#Distbrituion plot only used for numeric data (continous)
#histogram ----- distplot --only one column input
sns.distplot(tips.total_bill)
```

```python
sns.distplot(tips.total_bill,kde=False) #distplot without kde
```

```python
#joint plots 
#kind =scatter,hex,reg
# combination of histogram and scatter plot
sns.jointplot(x=tips.total_bill,y=tips.tip,kind = 'hex')
```

```python
#joint plots 
# combination of histogram and scatter plot
sns.jointplot(x=tips.total_bill,y=tips.tip,kind = 'reg')
```

```python
#controlling grids
iris=sns.load_dataset("iris")
```
